```r
df.v1.e <- read.csv('results//test_event_results_v1.csv')
df.v1.f <- read.csv('results/test_frame_results_v1.csv')
df.v2.e <- read.csv('results//test_event_results_v2.csv')
df.v2.f <- read.csv('results//test_frame_results_v2.csv')

df.v3.e <- read.csv('results//test_event_results_v3.csv')
df.v3.f <- read.csv('results//test_frame_results_v3.csv')
```

```r
library(ggplot2)
library(EnvStats)
```

```r
conductPaired <- function(df1, df2, col.name, alternative, fcn, fcn.string, showPlot = FALSE, ...){
    x <- df.v1.f[,col.name]
    y <- df.v2.f[,col.name]
    print(fcn(x) - fcn(y))

    t <- twoSamplePermutationTestLocation(
        x, y,
        alternative = alternative,
        fcn = fcn.string,
        paired = TRUE,
        ...)
    df <- data.frame(x = x, y = y)
    if (showPlot){
        boxplot(df)
    }
    if (alternative == 'greater'){
        print('Is X > Y  likely?')        
    } else if (alternative == 'less'){
        print('Is X < Y  likely?')
    }else {
        print('Is |X - Y| != 0  likely?')
    }
    print(t$p.value < 0.05)

}
```

```r
conductPaired(df.v1.f, df.v2.f, 'left_turn_recall', 'greater', median, 'median')
conductPaired(df.v1.f, df.v2.f, 'left_turn_recall', 'greater', mean, 'mean', showPlot = TRUE)
```

```r
conductPaired(df.v1.f, df.v2.f, 'left_turn_precision', 'less', median, 'median')
conductPaired(df.v1.f, df.v2.f, 'left_turn_precision', 'less', mean, 'mean', showPlot = TRUE)
```

```r
conductPaired(df.v1.f, df.v2.f, 'right_turn_precision', 'greater', median, 'median')
conductPaired(df.v1.f, df.v2.f, 'right_turn_precision', 'greater', mean, 'mean', showPlot = TRUE)
```

```r
conductPaired(df.v1.f, df.v2.f, 'right_turn_recall', 'greater', median, 'median')
conductPaired(df.v1.f, df.v2.f, 'right_turn_recall', 'greater', mean, 'mean', showPlot = TRUE)
```
